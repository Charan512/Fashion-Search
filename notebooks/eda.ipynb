# Fashionpedia Dataset — Exploratory Data Analysis

Explore the Fashionpedia dataset before indexing: image statistics, label distributions, color analysis.


```python
import sys
sys.path.insert(0, '..')  # project root

import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from part_a_indexer.dataset_processor import load_fashionpedia

# Load a small sample
images, metadata = load_fashionpedia(subset_size=200)
print(f'Loaded {len(images)} images')
```

## Image Size Distribution

```python
widths = [img.width for img in images]
heights = [img.height for img in images]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(widths, bins=30, color='#2ECC71', edgecolor='black')
axes[0].set_title('Width Distribution')
axes[0].set_xlabel('Pixels')
axes[1].hist(heights, bins=30, color='#F39C12', edgecolor='black')
axes[1].set_title('Height Distribution')
axes[1].set_xlabel('Pixels')
plt.tight_layout()
plt.show()
print(f'Width  — min: {min(widths)}, max: {max(widths)}, mean: {np.mean(widths):.0f}')
print(f'Height — min: {min(heights)}, max: {max(heights)}, mean: {np.mean(heights):.0f}')
```

## Sample Images

```python
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flatten()):
    if i < len(images):
        ax.imshow(images[i])
        ax.set_title(metadata[i]['image_id'], fontsize=8)
    ax.axis('off')
plt.suptitle('Sample Fashionpedia Images', fontsize=14)
plt.tight_layout()
plt.show()
```

## Zero-Shot Attribute Extraction on Sample

```python
from part_a_indexer.attribute_extractor import AttributeExtractor

extractor = AttributeExtractor(device='cpu')

# Extract attributes from first 5 images
for i, img in enumerate(images[:5]):
    attrs = extractor.extract_all_attributes(img)
    print(f'\nImage {i+1} ({metadata[i]["image_id"]}):')
    print(f'  Colors:    {attrs["primary_colors"]}')
    print(f'  Clothing:  {attrs["clothing_items"][:3]}')
    print(f'  Formality: {attrs["formality_score"]:.2f}')
    print(f'  Setting:   {attrs["setting"]}')
    print(f'  Style:     {attrs["style_category"]}')
```
